###Problema 1

####**Solución:**

#####**- Linealidad:**

  >Principio de homogeneidad: \\

 $\mathcal{L}\{\alpha f(x)\} = \alpha f(x)\tan{(2w_cx)} = \alpha \mathcal{L}\{f(x)\}$
 >Principio de superposición: \\

  $\mathcal{L}\{f(x)+g(x)\} = (f(x)+g(x))\tan{(2w_cx)} = f(x)\tan{(2w_cx)}+g(x)\tan{(2w_cx)} =\mathcal{L}\{f(x)\} + \mathcal{L}\{g(x)\}$

  Como se cumplen ambos principios el sistema es lineal

##### **- Causalidad:**
El sistema es causal debido a que la salida solo depende de los valores presentes, si $x =x_0$, la salida estaría dada por:\
$f(x_0) \tan(2w_cx_0)$\
 Lo cual solo depende del valor presente $x_0$

#####** - Estabilidad:**\
  Si $x = \frac{1}{4w_c\pi}$ la salida del sistema queda de la forma: \

 $y(\frac{1}{4w_c\pi}) = f(\frac{1}{4w_c\pi}) \tan{(\frac{\pi}{2})}$ \
 y $\tan{(\frac{\pi}{2})} = \infty$ por lo que para una entrada acotada se obtiene una salida no acotada, lo que implica que el sistema es inestable
 ##### **- Invariancia:**
 Para que el sistema sea invariante se debe cumplir:\
 $\mathcal{L}\{f(x)\} = y(x)$\
 $\mathcal{L}\{f(x-\xi)\} = y(x-\xi)$\
 Sin embargo para el sistema dado:\
$\mathcal{L}\{f(x-\xi)\} = f(x-\xi)\tan{(2w_cx)} \not= y(t-\xi)$\
Por lo tanto el sistema es variante
 ##### **- Invertibilidad:**
 Como $y(x)= f(x)\tan{(2w_cx)}$, entonces $\mathcal{L}^{-1}\{y(x)\} = \frac{y(x)}{\tan{(2w_cx)}}$ y buscamos $\mathcal{L}^{-1}\{\mathcal{L}\{f(x)\}\} = f(x)$
 \
 Reemplazando obtenemos: \
 $\mathcal{L}^{-1}\{f(x)\tan{(2w_cx)}\} = \frac{f(x)\tan{(2w_cx)}}{\tan{(2w_cx)}}=f(x)$\
 Por lo tanto el sistema es invertible

$\Phi (t) = e^{At}$ con $A = \begin{bmatrix}{0}&{1}\\{-a}&{0} \end{bmatrix}$.
Lo anterior se puede escribir como: \
$$\Phi(t) = e^{At} = I + At + \frac{A^2t^2}{2!}+ ....$$ \ 
Al evaluar $A$ en sus diferentes potencias obtenemos: \

$$A^2 = \begin{bmatrix}{-a}&{0}\\{0}&{-a} \end{bmatrix} = -a I$$ \
$$A^3 = -aA$$
$$A^4 = a^2 I ...$$ \
Por lo que $\Phi(t)$ queda: \
\
$$\Phi (t) = I(1-\frac{at^2}{2!}+a^2\frac{t^4}{4!}+...) + A(t-a\frac{t^3}{3!}+a\frac{t^5}{5!}) = Iw_1(t) + A w_2(t)$$ \
$$\Phi(t) = \begin{bmatrix}{w_1(t)}&{w_2(t)}\\{-aw_2(t)}&{w_1(t)} \end{bmatrix}$$
